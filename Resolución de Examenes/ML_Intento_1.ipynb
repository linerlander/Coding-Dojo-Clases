```python
import os
import seaborn as sns
output_folder = 'Gr√°ficos_Boxplots'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_columns:
    fig, ax = plt.subplots(figsize=(6, 4))
    sns.boxplot( data=df, x=col, ax=ax, color='skyblue')
    ax.set_title(f'Boxplot de {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    # Ajustar el dise√±o
    plt.tight_layout()

    output_path = os.path.join(output_folder, f'{col}_boxplot.png')
    fig.savefig(output_path)
    # Cerrar la figura para evitar problemas de memoria
    plt.close(fig)
print(f"Gr√°ficos de boxplots guardados en la carpeta '{output_folder}'.")

```

```python
output_folder = 'Gr√°ficos_Outliers_Tratados'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_tratado = df.copy()

for col in numeric_columns:
   
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_tratado[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

    fig, ax = plt.subplots(figsize=(6, 4))

    sns.boxplot(data=df_tratado, x=col, ax=ax, color='lightgreen')

    ax.set_title(f'Boxplot Tratado de {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    plt.tight_layout()

    output_path = os.path.join(output_folder, f'{col}_tratado_boxplot.png')
    fig.savefig(output_path)
    plt.close(fig)

df_tratado.to_csv('DataFrame_tratado.csv', index=False)

print(f"Gr√°ficos de boxplots tratados guardados en la carpeta '{output_folder}'.")
```
