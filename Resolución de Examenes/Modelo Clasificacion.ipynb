```python
output_folder = 'Gr√°ficos_Outliers_Tratados'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

numeric_columns = ['IMC','Promedio_nivel_glucosa']
df_tratado = df.copy()

for col in numeric_columns:
   
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_tratado[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

    fig, ax = plt.subplots(figsize=(6, 4))

    sns.boxplot(data=df_tratado, x=col, ax=ax, color='lightgreen')

    ax.set_title(f'Boxplot Tratado de {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    plt.tight_layout()

    output_path = os.path.join(output_folder, f'{col}_tratado_boxplot.png')
    fig.savefig(output_path)
    plt.close(fig)

df_tratado.to_csv('DataFrame_tratado.csv', index=False)

print(f"Gr√°ficos de boxplots tratados guardados en la carpeta '{output_folder}'.")
```

**4.1. Variables dummy y Pipeline**


```python
df['Estados_civil'].value_counts()
```

```python
columns = ['Genero','Estados_civil','Zona_residencia']
```

```python
for col in columns:
    dummies = pd.get_dummies(df_tratado[col], drop_first = False, dtype = float)
    dummies.columns = [f'{col}_{c}' for c in dummies.columns]
    df_tratado = pd.concat([df_tratado, dummies], axis=1)
    df_tratado.drop(columns=col, inplace=True)

```

```python
df_tratado.info()
```

```python
df_tratado[['Genero_mujer','Genero_other','Genero_hombre']]
```

>**Observaci√≥n**
>- Se puede observar que se tiene se ha creado variables dummy y asi tener valores n√∫mericos y no categoricos y son las columnas: `Estados_civil`, `Zona_residencia`y `Genero`

*Pipeline*

```python
from sklearn.compose import make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder, StandardScaler
from sklearn.compose import make_column_transformer
```

```python
cat_columnas_nominal = ['Tipo_trabajo','Flag_fumador']
columns_numericas = df_tratado.select_dtypes(exclude = 'object').columns.to_list()
```

```python
# Transformadores nominal
one_hot_encoder = OneHotEncoder()

#ordinal_encoder = OrdinalEncoder()

standard_scaler = StandardScaler()

imputer_categoricas = SimpleImputer(strategy='most_frequent')

imputer_num = SimpleImputer(strategy='mean')
```

```python
col_transformer = make_column_transformer(
    (make_pipeline(imputer_categoricas, one_hot_encoder), cat_columnas_nominal),
    (make_pipeline(imputer_num, standard_scaler), columns_numericas),
    remainder='passthrough'
)
```

```python
col_transformer
```

```python
col_transformer.fit(df_tratado)
```

```python
aux = col_transformer.transform(df_tratado)
df_procesado = pd.DataFrame(aux,columns=col_transformer.get_feature_names_out())
df_procesado.info()
```

```python
df_procesado.columns = df_procesado.columns.str.split('__').str[-1]
df_procesado.columns.to_list()
```

```python
df_procesado
```

**`4.2. Divisi√≥n de Muestra Train y Test`**

El objetivo de esta fase es poder divir en dos partes: el conjunto de entrenamiento (train) y el conjunto de prueba (test). Estas divisiones se utilizan para entrenar y evaluar los modelos.

- `|Train:|` El conjunto de entrenamiento se utiliza para entrenar el modelo de aprendizaje autom√°tico. Es aqu√≠ donde el modelo "aprende" los patrones y relaciones en los datos para poder hacer predicciones o clasificaciones.

- `|Test:|` El conjunto de prueba se utiliza para evaluar el rendimiento del modelo en datos no vistos durante el entrenamiento. Es una medida objetiva de la capacidad del modelo para generalizar y realizar predicciones precisas en nuevos datos.

**`6. Uso de Smote para balanceo`**

El objetivo de esta fase es poder balancear nuestra da, SMOTE genera nuevas muestras sint√©ticas para la clase minoritaria, lo que ayuda a balancear las clases.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Funci√≥n para graficar la matriz de confusi√≥n
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Matriz de Confusi√≥n - {model_name}')
    plt.xlabel('Predicciones')
    plt.ylabel('Valores Reales')
    plt.show()

# Graficar la matriz de confusi√≥n para Regresi√≥n Log√≠stica
plot_confusion_matrix(y_test, y_pred_logreg, "Regresi√≥n Log√≠stica")
```

- **Descripci√≥n de los valores**

    - **6270**: `Verdaderos negativos (TN)`. El modelo predijo "0" correctamente, y el valor real era "0".
    - **1998**: `Falsos positivos (FP)`. El modelo predijo "1", pero el valor real era "0".
